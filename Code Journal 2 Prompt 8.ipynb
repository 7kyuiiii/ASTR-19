```python
import numpy as np
import matplotlib.pyplot as plt

# 1) Create x from 0 to 1 in 100 steps
x = np.linspace(0, 1, 100)

# 2) Define functions that return sin(x) and cos(x)
def sin_func(x):
    return np.sin(x)

def cos_func(x):
    return np.cos(x)

# Compute y values
y_sin = sin_func(x)
y_cos = cos_func(x)

# Create multipanel plot (1 row, 2 columns)
fig, axs = plt.subplots(1, 2)

# Left panel: sin(x)
axs[0].plot(x, y_sin)
axs[0].set_title("sin(x)")

# Right panel: cos(x)
axs[1].plot(x, y_cos)
axs[1].set_title("cos(x)")

plt.tight_layout()
plt.show()
```
