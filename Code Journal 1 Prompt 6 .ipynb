```python
# Sin(x) and Cos(x) Table
# This notebook defines functions to compute sin(x) and cos(x).
# It then tabulates these functions for values of x between 0 and 2.
# Finally, it prints the first 10 values of x, sin(x), and cos(x).

import math

def my_sin(x):
    """
    Returns the sine of x.
    Input: x (float) - angle in radians
    Output: sin(x) (float)
    """
    return math.sin(x)

def my_cos(x):
    """
    Returns the cosine of x.
    Input: x (float) - angle in radians
    Output: cos(x) (float)
    """
    return math.cos(x)

# Number of points
num_points = 1000

# Range
x_start = 0.0
x_end = 2.0

# Step size
step = (x_end - x_start) / (num_points - 1)

# Lists to store results
x_values = []
sin_values = []
cos_values = []

for i in range(num_points):
    x = x_start + i * step
    x_values.append(x)
    sin_values.append(my_sin(x))
    cos_values.append(my_cos(x))

    print("x\t\tsin(x)\t\tcos(x)")
print("------------------------------------")

for i in range(10):
    print(f"{x_values[i]:.6f}\t{sin_values[i]:.6f}\t{cos_values[i]:.6f}")


```
